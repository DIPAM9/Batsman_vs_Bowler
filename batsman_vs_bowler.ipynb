# A function which can compare a batsman and a bowler and tell which one of them is a better T20 player


```python
def batsman_vs_bowler(batsman,bowler):
    a=data.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
    b=data.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)>300
    c=a[b].index.tolist()
    mask1=data[data['batsman'].isin(c)]
    mask1_runs=mask1.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
    mask1_balls=mask1.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False)
    mask1_strike=((mask1_runs/mask1_balls)*100).sort_values(ascending=False)
    x=data.groupby('bowler')['total_runs'].count()
    y=data.groupby('bowler')['total_runs'].count()>100
    z=x[y].index.tolist()
    mask2=data[data['bowler'].isin(z)]
    mask2_runs=mask2.groupby('bowler')['total_runs'].sum().sort_values(ascending=False)
    mask2_balls=mask2.groupby('bowler')['total_runs'].count().sort_values(ascending=False)
    mask2_overs=(mask2_balls/6)
    mask2_economy=(mask2_runs/mask2_overs).sort_values()
    if((batsman in c)&(bowler in z)):
        p=(mask1_strike[batsman]-mask1_strike.mean())/statistics.stdev(mask1_strike)
        q=(mask2_economy.mean()-mask2_economy[bowler])/statistics.stdev(mask2_economy)
        if(p>q):
            print("The better IPL player: ",batsman)
        else:
            print("The better IPL player: ",bowler)
    else:
        print(" ",batsman,"is not a batsman &",bowler, "is not a bowler")
```
